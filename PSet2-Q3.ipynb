```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

np.random.seed(1234)
```

```python
u = np.random.standard_normal(1025)
v = np.random.standard_normal(1025)

x = np.zeros(1025)
y = np.zeros(1025)

for i in range(0,1025):
    x[i] = x[i-1] + u[i]
    y[i] = y[i-1] + v[i]

x = x[-25:]
y = y[-25:]

x = x.reshape((-1,1))

model = LinearRegression().fit(x,y)

print("O intercepto da regressão é ",model.intercept_,", o coeficiente é ",
      str(model.coef_[0]), " e o R² é ",model.score(x,y))
```

## b)

```python
x_values = []
y_values = []
intercepts = []
slopes = []
r_sqs = []
```

```python
for j in range(0,2000):
    u = np.random.standard_normal(1025)
    v = np.random.standard_normal(1025)
    
    x = np.zeros(1025)
    y = np.zeros(1025)
    
    for i in range(0,1025):
        x[i] = x[i-1] + u[i]
        y[i] = y[i-1] + v[i]
    
    x = x[-25:]
    y = y[-25:]
    
    x_values.extend(x)
    y_values.extend(y)
    
    x = x.reshape((-1,1))
    
    model = LinearRegression().fit(x,y)
    
    intercepts.append(model.intercept_)
    slopes.append(model.coef_[0])
    r_sqs.append(model.score(x,y))
```

```python
data_names = ['x_values','y_values','intercepts','slopes','r_sqs']
data = [x_values,y_values,intercepts,slopes,r_sqs]

fig, axs = plt.subplots(3,2, figsize=(10,10))
axs = axs.ravel()

for i in range(len(data)):
    axs[i].hist(data[i], bins = 'auto')
    axs[i].set_title("Histogram of " + data_names[i])
    axs[i].axvline(np.mean(data[i]), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    axs[i].text(np.mean(data[i])*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(np.mean(data[i])))
    
plt.tight_layout()
plt.show()
```
